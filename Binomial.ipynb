# The Binomial Distribution

Bionmial Distributions represent the probability mass function of outcomes like coin flips.

```python
%matplotlib inline
import matplotlib.pyplot as plt

from functools import reduce

#where n is the total number of trails and k is the number of successes

n = 10
k = 6
```

```python
n_product_k = [i for i in range(k+1, n+1)]
difference_product = [i for i in range(1, n-k+1)]
```

This formula for the binomial distribution $(\frac{n}{k}) \frac{n!}{k!(n-k)!$ simplifies, as many textbooks and articles note.
These are the factors of the final simplified numerator for the given example.

Factorial (signified by !) is simplied to a product over a range, e.g. $\Pi_{3}^{5}$  

```python
numerator = reduce(lambda x,y: x*y, n_product_k)
denominator = reduce(lambda x,y: x*y, difference_product)
```

```python
def successes(n, k):
    numerator = reduce(lambda x,y: x*y, [i for i in range(k+1, n+1)])
    denominator = reduce(lambda x,y: x*y, [i for i in range(1, n-k+1)])
    return numerator/denominator
```

```python
distribution = [successes(n, i) for i in range(n)]
```

```python
n = 10
eta = 10

import numpy as np

#lin_x_space = np.linspace(1,100, 1)
 
k = n+1
for m in range(1, k):
    plt.figure(m)
    n = m*eta
    distribution = [successes(n, i) for i in range(n)]
                    
    #print((k-1)**2/n, len(lin_x_space[::((k-1)**2)/n]), len(distribution))
    lin_x_space = [i for i in range(n)]    
    
    total = sum(distribution)
    remaining = [total-distribution[i] for i in range(n)]
    #plt.subplot(m/5, m%5, m+1)
    plt.plot(lin_x_space, distribution, "b", label="{0}".format(m))
    plt.legend(loc="best")
plt.show()
```
